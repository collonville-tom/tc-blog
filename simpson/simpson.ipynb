```python
import numpy as np

import matplotlib.pyplot as plt


```

```python
def generateSet(prototype,nbrEchantillon,coef):
    rand_value=np.random.randn(len(prototype),len(prototype[0]))/coef
    #print(rand_value)
    rand_set=prototype+rand_value
    if nbrEchantillon == 0 :
        return prototype
    else:
        return np.concatenate((rand_set,generateSet(prototype,nbrEchantillon-1,coef)))
```

```python
evidenceBasedMedGravePattern=np.array([[0.20]])
evidenceBasedMedGrave=generateSet(evidenceBasedMedGravePattern,200,25)
evidenceBasedMedPasGravePattern=np.array([[0.80]])
evidenceBasedMedPasGrave=generateSet(evidenceBasedMedPasGravePattern,200,25)

fakeMedGravePattern=np.array([[0.01]])
fakeMedGrave=generateSet(fakeMedGravePattern,25,50)
fakeMedPasGravePattern=np.array([[0.30]])
fakeMedPasGrave=generateSet(fakeMedPasGravePattern,2000,50)


print("evidenceBasedMedGrave",np.mean(evidenceBasedMedGrave)*100)
print("evidenceBasedMedPasGrave",np.mean(evidenceBasedMedPasGrave)*100)
print("fakeMedGrave",np.mean(fakeMedGrave)*100)
print("fakeMedPasGrave",np.mean(fakeMedPasGrave)*100)


print("evidenceBasedMed",np.mean(np.concatenate((evidenceBasedMedGrave,evidenceBasedMedPasGravePattern)))*100)
print("fakeMed",np.mean(np.concatenate((fakeMedGrave,fakeMedPasGrave)))*100)
```

```python
fig = plt.figure(1,figsize=(12,12))

plt.plot(np.ones(201),evidenceBasedMedPasGrave,"b.")
plt.plot(np.ones(2001),fakeMedPasGrave,"r.")
plt.plot(np.ones(201)*2,evidenceBasedMedGrave,"y.")
plt.plot(np.ones(26)*2,fakeMedGrave,"g.")
plt.xlabel('Gravit√©')
plt.ylabel('Retablissement')
```
